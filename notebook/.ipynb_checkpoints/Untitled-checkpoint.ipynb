```python
def check(s):
    n = len(s)
    for i in range(1, n+1):
        if n % i == 0:
            stop = True
            for j in range(n / i):
                if s[j*i:(j+1)*i] != s[:i]:
                    stop=False
                    break
            if stop:
                return n/i
```
